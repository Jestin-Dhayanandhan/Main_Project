```python
import tkinter as tk

board = [[9, 0, 6, 0, 1, 3, 0, 0, 8],
        [0, 5, 8, 0, 0, 0, 0, 9, 0],
        [0, 3, 0, 0, 0, 0, 0, 1, 0],
        [0, 6, 0, 8, 0, 0, 9, 2, 0],
        [0, 0, 3, 4, 0, 9, 1, 0, 0],
        [0, 4, 9, 0, 0, 6, 0, 3, 0],
        [0, 9, 0, 0, 0, 0, 0, 8, 0],
        [0, 1, 0, 0, 0, 0, 6, 7, 0],
        [4, 0, 0, 9, 6, 0, 3, 0, 1]]

#TO find the empty position from the board:
def find_empty_position(board):
    for i in range(9):
        for j in range(9):
            if board[i][j]==0:
                return i,j
    return False

#To check number is repeating in row or column or #associated 3X3 matrix:
def is_valid(board,num,position):
    for i in range(9):
        if board[position[0]][i]==num:
            return False
    for j in range(9):
        if board[j][position[1]]==num:
            return False
    x,y=(position[0]//3)*3,(position[1]//3)*3    

    for i in range(3):
        for j in range(3):
            if board[x+i][y+j]==num:
                return False
    return True

#To solve suduko problem:
def slove(board):
    position=find_empty_position(board)

    if position== False:
        return True

    for num in range(1,10):
        if is_valid(board,num,position):
            board[position[0]][position[1]]=num
            if slove(board):
                return True
    board[position[0]][position[1]]=0
    return False

#To draw table:
def draw_suduko(board):
    for i in range(1,9):
        line_width=3 if i%3==0 else 1

        #Horizontal_line
        canvas.create_line(0,i*50,450,i*50,width=line_width)
        #Vertitical line
        canvas.create_line(i*50,0,i*50,450,width=line_width) 

    for i in range(9):
        for j in range(9):
            canvas.create_text(j*50+25,i*50+25,text=str(board[i][j]),font=('Helvetica', 16))            

#Solving
slove(board)

#Displaying in the table
root=tk.Tk()
root.title('SUDUKO SOLVER')

canvas=tk.Canvas(root,width=450,height=450)
canvas.pack()

draw_suduko(board)

root.mainloop()
```
